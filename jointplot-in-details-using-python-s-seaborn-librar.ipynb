```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.set_style('dark')
sns.jointplot(x='sepal_length',y='sepal_width', data= iris,s=100)
plt.show()
```

```python
sns.set_style('dark')
g = sns.jointplot(x='sepal_length',y='sepal_width',
              data = iris,s = 100,
                 marginal_ticks=True,
             marginal_kws = dict(bins = 25, fill = False))
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", 'Sepal Width', fontsize=14)
plt.show()
```

```python
sns.set_style('dark')
g = sns.jointplot(x="sepal_length",y='sepal_width',
                 kind='reg', data= iris,
                 marginal_ticks=True,
                 marginal_kws=dict(bins= 24, fill = False))
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", 'Sepal Width', fontsize=14)
plt.show()
```

```python
sns.set_style('dark')
g = sns.jointplot(x="sepal_length", y='sepal_width',data = iris,
                 hue = 'species', marginal_ticks=True)
g.fig.set_size_inches(8,8)
g.set_axis_labels('Sepal Length', 'Sepal Width',fontsize=14)
plt.show()
```

```python
sns.set_style('dark')
g = sns.jointplot(x='sepal_length', y='sepal_width', 
                  hue = 'species',
                 kind = 'kde', marginal_ticks=True,
                 data = iris)
g.fig.set_size_inches(8,8)
g.set_axis_labels('Sepal Length','Sepal Width')
plt.show()
```

```python
sns.set_style('dark')
g = sns.jointplot(x='sepal_length', y='sepal_width', 
                 data = iris, kind = 'hex',
                 marginal_ticks=True)

g.fig.set_size_inches(8,8)
g.set_axis_labels('Sepal Length','Sepal Width',fontsize=14)
plt.show()
```

```python
sns.set_style('dark')
g = sns.jointplot(x='sepal_length', y='sepal_width',                 
                 kind = 'kde', marginal_ticks=True,
                 data = iris)
g.fig.set_size_inches(8,8)
g.set_axis_labels('Sepal Length','Sepal Width')
plt.show()
```

```python
sns.set_style('dark')
g = sns.jointplot(x='sepal_length', y='sepal_width',                 
                 marginal_ticks=True,s = 100,
                 data = iris)
g.plot_joint(sns.kdeplot, color = 'red')
g.fig.set_size_inches(8,8)
g.set_axis_labels('Sepal Length','Sepal Width')
plt.show()
```
